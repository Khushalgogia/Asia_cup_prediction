```python
model = Sequential()
model.add(Dense(256,activation = 'relu',input_dim = 34))
model.add(BatchNormalization())
model.add(Dense(256,activation = 'relu'))
model.add(BatchNormalization())
model.add(Dense(1, activation = 'linear'))
model.compile(optimizer = 'adam',loss = 'mean_squared_error',metrics = 'mse')
model.summary()
```

```python
#Linear_Regression

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

lr = LinearRegression()
model = lr.fit(xTrain,yTrain)
train_acc0 = r2_score(yTrain,model.predict(xTrain))
test_acc0 = r2_score(yTest,model.predict(xTest))
mse0 = mean_squared_error(yTest,model.predict(xTest))

print('Mean Squared Error', mse0)

print('Train SCore: ',train_acc0)
print('Test SCore: ',test_acc0)
```

```python
# Decision Tree


from sklearn.tree import DecisionTreeRegressor
df = DecisionTreeRegressor()
model2 = df.fit(xTrain,yTrain)
pred1 = model2.predict(xTest)
train_acc = r2_score(yTrain,model2.predict(xTrain))
test_acc = r2_score(yTest,model2.predict(xTest))
mse1 = mean_squared_error(yTest,model2.predict(xTest))


print('Train SCore: ',train_acc)
print('Test SCore: ',test_acc)
print('Mean Squared Error', mse1)
```

```python
# Random Forest


from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model1 = rf.fit(xTrain,yTrain)
pred = model1.predict(xTest)
train_acc1 = r2_score(yTrain,model1.predict(xTrain))
test_acc1 = r2_score(yTest,model1.predict(xTest))
mse2 = mean_squared_error(yTest,model1.predict(xTest))

print('Mean Squared Error', mse2)
print('Train SCore: ',train_acc1)
print('Test SCore: ',test_acc1)
```

```python
#Bagging

from sklearn.ensemble import BaggingRegressor
bg = BaggingRegressor()
model4 = bg.fit(xTrain,yTrain)
train_acc3 = r2_score(yTrain,model4.predict(xTrain))
test_acc3 = r2_score(yTest,model4.predict(xTest))
mse4 = mean_squared_error(yTest,model4.predict(xTest))

print('Mean Squared Error', mse4)
print('Train SCore: ',train_acc3)
print('Test SCore: ',test_acc3)
```

```python
# XGBoost

from xgboost import XGBRegressor
xgb = XGBRegressor()
model3 = xgb.fit(xTrain,yTrain)
train_acc2 = r2_score(yTrain,model3.predict(xTrain))
test_acc2 = r2_score(yTest, model3.predict(xTest))
mse3 = mean_squared_error(yTest,model3.predict(xTest))

print('Mean Squared Error', mse3)

print('Train SCore: ',train_acc2)
print('Test SCore: ',test_acc2)
```
